```python
import pandas as pd
```

```python
df=pd.read_csv("D_2 1.csv", sep=";" )
```

```python
df.head()
```

```python
df["bejelentve"]=pd.to_datetime(df["bejelentve"])
df["Csoporton"]=pd.to_datetime(df["Csoporton"])
df["Kollégán"]=pd.to_datetime(df["Kollégán"])
```

```python
df["bejelentve"]=df["bejelentve"].dt.strftime("%Y. %m. %d. %H:%M:%S")
df["Csoporton"]=df["Csoporton"].dt.strftime("%Y. %m. %d. %H:%M:%S")
df["Kollégán"]=df["Kollégán"].dt.strftime("%Y. %m. %d. %H:%M:%S")
```

```python
df.columns=["inc_id","letrejott","folyamatban","felveve"]
```

```python
df2=df[["letrejott","folyamatban","felveve","inc_id"]]
```
